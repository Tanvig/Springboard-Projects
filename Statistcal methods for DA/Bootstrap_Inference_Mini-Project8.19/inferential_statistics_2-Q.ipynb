## Introduction

In the previous frequentist mini-projects, you did frequentist calculations to perform inference from a sample of data. Such inference relies on theory largely developed from the 19th-Century onwards that is subject to certain assumptions or theoretical limits. These are fine if those assumptions hold for the particular case you're working on, and what you want to do has a known theoretical distribution (for example the mean of a sampling distribution that we looked at in the previous mini-project.)

In this mini-project, you'll use the same medical charge data you used in the frequentist inference mini-project, but this time you'll make inferences about the population using bootstrapping (ie. simulating repeated re-runs of an experiment.) If frequentism is about using assumptions and theoretical results to calculate what we expect to happen were an experiment to be run again and again and again, then bootstrapping is about using computing power to essentially re-run the sample draw again and again and again to see what actually happens.

__Q:__ Use bootstrap sampling to estimate the same 95% confidence interval lower limit as before.

__A:__

```python
np.random.seed(47)
N_rep = 10000

charges = np.array(med_charges["charges"])
bs_replicates = np.empty(N_rep)

for i in range(N_rep):
    bs_replicates[i] = np.mean(np.random.choice(charges, size = len(charges)))

#Since we need the lower limit and it is one tailed

conf_int = np.percentile(bs_replicates, (5))
conf_int
```

If you performed 10000 replicates immediately after setting the random seed to 47, you should get the value 12724 here, which compares very well with the value 12725 obtained using the _t_-distribution confidence interval previously. It is a most pleasant result to see the predictions of classical frequentist theory match with results that are now possible through the number-crunching ability of computers.

__Q:__ Calculate the 95% confidence interval for the difference between the standard deviations of insurance and non-insurance claim charges (insured - non-insured). Calculate the differences over 10000 replicates. Plot the histogram of values and mark the locations of the percentiles. State the null and alternative hypothesis and comment on whether you would retain or reject the null hypothesis in this case and why.

__A:__

```python
# Null hypothesis : There is no difference between the standard deviations of insurance and non-insurance claim charges.
# Alternate hypothesis:  There is difference between the standard deviations of insurance and non-insurance claim charges.

# OR

# Null hypothesis : Std of insurance claim charges == Std of non-insurance claim charges.
# Alternate hypothesis:  Std of insurance claim charges != Std of non-insurance claim charges.
```

```python
np.random.seed(47)
N_rep = 10000

insured = med_charges["charges"][med_charges["insuranceclaim"] == 1]
uninsured = med_charges["charges"][med_charges["insuranceclaim"] == 0]

obs_diff = np.std(insured) - np.std(uninsured)
obs_diff
```

```python
bs_rep = np.empty(N_rep)

for i in range(N_rep):
    bs_insured = np.random.choice(insured, size = len(insured))
    bs_uninsured = np.random.choice(uninsured, size = len(uninsured))
    bs_rep[i] = np.std(bs_insured) - np.std(bs_uninsured)

conf_inv = np.percentile(bs_rep, (2.5, 97.5))
conf_inv
```

```python
plt.hist(bs_rep, bins = 25, density = True)
plt.axvline(conf_inv[0], color='red')
plt.axvline(conf_inv[1], color='red')
plt.axvline(np.mean(bs_rep), color='red', linestyle = "--")
plt.xlabel('Difference in Standard Deviations of Bootstrap replicates')
plt.ylabel('PDF')
plt.show()
```

```python
# As per our Null hypothesis, our difference between the std of insured and non-insured claim charges should be around 0.
# But from our confidence interval, we can see that 0 does not belong to the interval obtained [6681.572467  , 8502.74820416]
# Hence, we can reject the Null hypothesis
```

## Confidence interval and p-value

The confidence interval above is often a useful quantity to estimate. If we wish to limit our expected probability of making a Type I error (where we wrongly reject the null hypothesis, and it is, instead, true) to $\alpha$, the associated confidence interval is our estimate of the interval within which we expect the true population value to be found $100\times(1 - \alpha)$% of the time we do this test. In the above we performed bootstrap replicates to estimate the interval and reject the null hypothesis if this interval did not contain zero. You will sometimes see such an interval reported in the output of statistical functions.

The partner of the confidence interval is the p-value. The p-value and the confidence interval are linked through our choice of $\alpha$. The p-value tells us how likely it is, under the null hypothesis, to get an outcome at least as extreme as what was observed. If this fails to reach the level of our _pre-specified_ $\alpha$, we decide the null hypothesis is sufficiently unlikely to be true and thus reject it. To calculate this p-value via the bootstrap, we have to put ourselves in a position where we are simulating the null hypothesis being true and then calculate the fraction of times we observe a result at least as extreme as that actually observed.

Remember how, previously, you used the _t_-test to calculate the p-value for the observed difference between the means of insured and non-insured medical cases. We're now going to repeat this, this time using the bootstrap approach.

__Q:__ Perform a bootstrapped hypothesis test at the 5% significance level ($\alpha = 0.05$) to calculate the p-value of the observed difference between insurance and non-insurance charges, state your null and alternative hypotheses and whether you retain or reject the null hypothesis for the given significance level.

__A:__

```python
# Null hypothesis - There is no difference between insured and uninsured claim charges i.e 
#                   Mean of insured claim charges = Mean of uninsured claim charges

# Alternate hypothesis - There is a difference between insured and uninsured claim charges i.e 
#                   Mean of insured claim charges != Mean of uninsured claim charges
```

```python
np.random.seed(47)
N_rep = 10000

insured = med_charges["charges"][med_charges["insuranceclaim"] == 1]
uninsured = med_charges["charges"][med_charges["insuranceclaim"] == 0]

obs_mean_diff = np.mean(insured) - np.mean(uninsured)
obs_mean_diff
```

```python
# A 5% significance level ( alpha = 0.05) means (100 * (1 - 0.05)) = 95% confidence level 
# To calculate the p-value via the bootstrap, we have to put ourselves in a position where we are simulating 
# the null hypothesis being true and then calculate the fraction of times we observe a result at least as extreme
# as that actually observed.

# Hence, we need to simulate, according to the null hypothesis, that charges with insured and uninsured claims have same mean.

total_mean_charges = np.mean(med_charges["charges"])

shifted_insured = insured - np.mean(insured) + total_mean_charges
shifted_uninsured = uninsured - np.mean(uninsured) + total_mean_charges

(total_mean_charges, np.mean(shifted_insured), np.mean(shifted_uninsured))
```

```python
np.random.seed(47)
bs_reps = np.empty(N_rep)

for i in range(N_rep):
    bs_insured = np.mean(np.random.choice(shifted_insured, size = len(shifted_insured)))
    bs_uninsured = np.mean(np.random.choice(shifted_uninsured, size = len(shifted_uninsured)))
    bs_reps[i] = bs_insured - bs_uninsured
```

```python
p_val = np.sum(bs_reps >= obs_mean_diff)/len(bs_reps)
p_val
```

```python
# Here, p_value < alpha i.e. 0 < 0.05
# Hence, our Null hypothesis isn't true. It is very unlikely to get such an extreme observed mean difference by chance.
```
